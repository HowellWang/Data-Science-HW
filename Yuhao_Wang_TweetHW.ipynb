I use nltk package help me to clean these stop words, however, there are also some words needs to add into the stop_words set like ever, even and so on



```python
stop_words = set(nltk.corpus.stopwords.words('English'))
stop_words.update(['one', 'time', 'times','amp','ever', 'even', 'twitter', 'michelle','cant', 'joe', 'john', 'lewis', 'gonna', 'going','talk','say','year'])
stop_words
```

Although stopwords can help me filter lots of irrelevant words, there are still something need to deal with. In this situation, I define a helper function to address the problems like punctuaion and person name.

```python
#define data cleaning api
def clean(input):
    outcome= []
    tmp = input.split(" ")
    for i in range(len(tmp)-1):
        r='[â€™!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~]+'
        tmp[i] = re.sub(r, '', tmp[i])
        outcome.append(tmp[i])
        if 'RT' in tmp[i] or 'http' in tmp[i] or 'say' in tmp[i] or 'talk' in tmp[i] or '&amp;' in tmp[i] or 'binden' in tmp[i] \
        or 'time' in tmp[i]or 'Biden'in tmp[i] or 'barack' in tmp[i] or 'Barack' in tmp[i] or 'obama' in tmp[i] or 'Obama' in tmp[i] \
        or 'donald' in tmp[i] or 'Donald' in tmp[i] or 'trump' in tmp[i] or 'Trump' in tmp[i]:
            outcome.remove(tmp[i])
    myStr =''
    for i in outcome:
        myStr = myStr + str(i) + ' '
    return myStr
```

```python
sT = clean(sT)
trump_D = {}
for word in sT.split(" "):
    word = word.lower()
    if word not in stop_words:
        if len(word) > 2:
            if word in trump_D.keys():
                trump_D[word] += 1
            else:
                trump_D[word] = 1
```

```python
sO = clean(sO)
obama_D = {}
for word in sO.split(" "):
    word = word.lower()
    if word not in stop_words:
        if len(word) > 2:
            if word in obama_D.keys():
                obama_D[word] += 1
            else:
                obama_D[word] = 1
```

I choose the first 15th key value pairs after sorting in order to avoid the situation that the 10th paris include something I don't want.

```python
items_T = [(v, k) for k, v in trump_D.items()]
items_T.sort()
items_T.reverse()             
items_T = [(k, v) for v, k in items_T]
items_T = items_T[0:15]
items_T
```

```python
items_O = [(v, k) for k, v in obama_D.items()]
items_O.sort()
items_O.reverse()             
items_O = [(k, v) for v, k in items_O]
items_O = items_O[0:15]
items_O
```

```python
items_T.remove(items_T[8])
items_T.remove(items_T[8])
items_T = items_T[0:10]
items_T
```

```python
items_O.remove(items_O[2])
items_O.remove(items_O[6])
items_O.remove(items_O[9])
items_O = items_O[0:10]
items_O
```

```python
#Define a helper function to set the sentiment value
def sentimentValue(x):
    if x == 'rights' or x == 'civil' or x == 'attack' or x == 'new' or x == 'country':
        return 1
    elif x == 'slytherin' or x == 'fuck' or x == 'ill':
        return -1
    else:
        return 0
```

```python
t_totalV = 0
t_totalF = 0
for i in range(10):
    t_totalV += sentimentValue(items_T[i][0]) * items_T[i][1]
    t_totalF += items_T[i][1]
```

```python
t_sentimentScore = (t_totalV + 0.0) / t_totalF
t_sentimentScore
```

```python
o_totalV = 0
o_totalF = 0
for i in range(10):
    o_totalV += sentimentValue(items_O[i][0]) * items_O[i][1]
    o_totalF += items_O[i][1]
```

```python
o_sentimentScore = (o_totalV + 0.0) / o_totalF
o_sentimentScore
```
